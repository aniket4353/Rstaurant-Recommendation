```python
def recommend(name, cosine_similarities=cosine_similarities):
    recommend_restaurant = []

    # Check if the restaurant exists in df2_percent
    if name in df2_percent['name'].values:
        # Get the index of the restaurant by name directly from the df2_percent DataFrame
        idx = df2_percent[df2_percent['name'] == name].index[0]

        score_series = pd.Series(cosine_similarities[idx]).sort_values(ascending=False)
        top30_indexes = list(score_series.iloc[0:31].index)

        for each in top30_indexes:
            recommend_restaurant.append(list(df2_percent.index)[each])

        # Create a new DataFrame with specified columns using pd.DataFrame
        df2_new = pd.DataFrame(columns=['cuisines', 'Mean Rating', 'cost'])

        for each in recommend_restaurant:
            # Append data to df2_new using pd.DataFrame and .loc
            df2_new = pd.concat([df2_new, pd.DataFrame(df2_percent.loc[[each], ['cuisines', 'Mean Rating', 'cost']])])

        df2_new = df2_new.drop_duplicates(subset=['cuisines', 'Mean Rating', 'cost'], keep=False)
        df2_new = df2_new.sort_values(by='Mean Rating', ascending=False).head(10)

        print('TOP %s RESTAURANTS LIKE %s WITH SIMILAR REVIEWS: ' % (str(len(df2_new)), name))

        return df2_new
    else:
        print(f"Restaurant '{name}' not found in the dataset.")
        return None # or an empty DataFrame
```
